```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
covid = pd.read_csv('New_Data/Covid_new.csv')
covid.head()
```

```python
covid.info()
```

```python
data = covid.groupby("Continent")["TotalCases"].sum()
tempt_value = list(data.values)
tempt_name_of_continents = list(data.keys())
sns.set_theme(style="darkgrid")
titanic = sns.load_dataset("titanic")
plt.figure(figsize = (10, 7))
plt.pie(tempt_value, labels = tempt_name_of_continents, autopct='%1.2f%%')
plt.title("Phần trăm ca nhiễm COVID-19 của các châu lục trên thế giới")
plt.show() 
```

#### Châu Âu và Bắc Mĩ chiếm hơn 1/2 tỷ lệ về số ca nhiễm về nguyên do có lẽ người dân không chịu đeo khẩu trang và tuân theo các biện pháp phòng dịch của chính phủ. Châu Á cũng có tỷ lệ ca nhiễm cao do đông dân 

```python
data = covid.groupby("Continent")["TotalDeaths"].sum()
tempt_value = list(data.values)
tempt_name_of_continents = list(data.keys())
sns.set_theme(style="darkgrid")
titanic = sns.load_dataset("titanic")
plt.figure(figsize = (10, 7))
plt.pie(tempt_value, labels = tempt_name_of_continents, autopct='%1.2f%%')
plt.title("Phần trăm ca tử vong COVID-19 của các châu lục trên thế giới")
plt.show() 
```

#### Châu Âu, Bắc Mĩ, Nam Mĩ và Châu Á là đều có số ca tử vong cao

```python
tempt_data = covid.sort_values(by=["TotalCases"], ascending=False)[0:20]
tempt_data = tempt_data.sort_values(by=["TotalCases"], ascending=True)
plt.figure(figsize = (18, 8))
plt.barh(tempt_data["Country,Other"], tempt_data["TotalCases"])

for i, v in enumerate(tempt_data["TotalCases"]):
    plt.text(v, i - 0.1, str(v), fontsize = 9)
sns.set_theme(style="darkgrid")
titanic = sns.load_dataset("titanic")
plt.title("Số ca nhiễm của 20 nước cao nhất trên thế giới")
plt.show()
```

#### Ta thấy Hoa Kỳ là quốc gia có số ca nhiễm nhiều nhất trên thế giói

```python
totalcases = covid['TotalCases'].sum()
us_confirmed = covid[covid['Country,Other']=='USA']['TotalCases'].sum()
outside_US = totalcases - us_confirmed

plt.figure(figsize=(10, 6))
plt.barh('United States', us_confirmed)
plt.barh('Outside United States', outside_US)
plt.title('Số ca nhiễm của Hoa Kỳ so với thế giới', size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()
```

#### Ta có thể thấy số ca nhiễm của Hoa Kỳ khoảng 1/4 so với tổng các nước trên thế giới và bằng 1/5 so với thế giới

```python
totalcases = covid['TotalDeaths'].sum()
us_confirmed = covid[covid['Country,Other']=='USA']['TotalDeaths'].sum()
outside_US = totalcases - us_confirmed

plt.figure(figsize=(10, 6))
plt.barh('United States', us_confirmed)
plt.barh('Outside United States', outside_US)
plt.title('Số tử vong của Hoa Kỳ so với thế giới', size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()
```

#### Ta có thể thấy số tử vong của Hoa Kỳ khoảng 1/5 so với tổng các nước trên thế giới và bằng 1/6 so với thế giới

```python
plt.figure(dpi=1000)
sns.heatmap(np.round(covid.corr(),2),annot=True)
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.regplot(x='TotalCases',y='TotalDeaths',data=covid,color='red')
plt.xlabel('TotalCases')
plt.ylabel('TotalDeaths')
plt.show()
```

#### Ta có thể thấy TotalCases và TotalDeaths tỷ lệ thuận với nhau khi số ca dương tính càng tăng thì số ca tử vong cũng tăng theo

```python
plt.figure(figsize=(12,6),dpi=100)
sns.regplot(x='TotalCases',y='TotalRecovered',data=covid,color='green')
plt.xlabel('TotalCases')
plt.ylabel('TotalRecovered')
plt.show()
```

#### TotalCases và TotalRecovered quan hệ tuyến tính với nhau đều này có nghĩa là số ca dương tính càng tắng thì số ca phục hồi do đã bị bệnh trước đó cũng tăng

```python
plt.figure(figsize=(12,6),dpi=100)
sns.regplot(x='TotalCases',y='TotalTests',data=covid,color='blue')
plt.xlabel('TotalCases')
plt.ylabel('TotalTests')
plt.show()
```

#### TotalTests và TotalCases có tỷ lệ thuận với nhau. Điều này có nghĩa là khi càng Test nhiều thì sẽ càng phát hiện được nhiều ca nhiễm Covid-19 mới
